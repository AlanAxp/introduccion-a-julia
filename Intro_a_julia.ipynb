Ahora usando lo que hemos aprendido podemos implementarlo para poder aproximar el valor 
de $\pi$

La explicacion del siguiente codigo es analoga a la que se ve en el codigo de Python


```julia
#Aproximaremos el valor de pi con el metodo de MonteCarlo
function pi_mc_2d(shoots)
    inicio = 0.0
    for i in 1:shoots
        x = rand()
        y = rand()
        if x*x+y*y <= 1 
            inicio = inicio + 1
        end
    end
    return (inicio / shoots)*4
end
function lista_pi(A,N) # A es la long de lista que quermos, N = puntos arrojados al cuadradp
    pis = []
    for i in 1:A
        push!(pis,pi_mc_2d(N))
    end
    return pis
end

function prom_var_err(DAT)
    prom = 0.0
    cuadrados = 0.0
    N = length(DAT)
    for i in 1:N
        prom = prom + DAT[i]
        cuadrados = cuadrados + DAT[i]*DAT[i]   
    end
    xbar = prom/N #    Valor del promedio
    var = (cuadrados - N*xbar*xbar)/(N-1)  #Aquí calculamos la varianza
    Er = sqrt(var/N)
    return xbar,var,Er
end
```

```julia
#Ejecutando nuestro programa...


@time Pis = lista_pi(1000,40000); #bajar parametros para que su compu no tarde mucho
result = prom_var_err(Pis)


print(" El valor aprox de π es ", result[1]," con un error estandar de ", result[3])
```

## "Programacion orientada a objetos" en julia

En julia no existe el paradigma tan cual como lo es la programacion orientada o objetos, no existe el concepto de clase tal cual.

Julia tiene algo que se llama estructuras, asi como en C y Rust y tienen un comportamiento al que podriamos considerar "analogo" al de las clases.

```julia
# una clase sera una estructura, mutable para que sus campos 
# puedan ser mutables

g = 9.8 ::Float64
Fy = 0.0 ::Float64

mutable struct Pelotita
    y::Float64
    v_y::Float64
    m::Float64
end

# Para definir un metodo, tenemos que pasar la estructura como argumento
function calcularFuerza(P::Pelotita)
    Fy = - P.m * g
    return Fy
end

# Creamos el método que describe el movimiento
function muevete(P::Pelotita, dt)
    P.y += P.v_y * dt
    P.v_y += dt * Fy / P.m
    return P.y
end

```

```julia

# Instanciar el objeto, con los parametros definidos en el constructor

bolita = Pelotita( 0.0, 3.0, 5.0 )

# Llamamos al médoto para calcular el peso
Fy = calcularFuerza(bolita)

# Creamos una iteracion para un intervalo de tiempo 
t = 0.0
Delta_t = 0.01

T = []
Y = []

while t < 1.0
    y = muevete(bolita, Delta_t)
    t += Delta_t
    push!(T, t)
    push!(Y, y)
end
```

```julia
plot(T, Y,  title = "T vs Y", marker =:o
        , xlabel = "T", ylabel = "Y", size = (500, 500), label = "Linea")
```
